```python
ulta = pd.read_csv('ulta.csv')
ulta.sample(5)
```

### Step 1: Find/remove the NAs!
If a row is NA for review info, it means there have been 0 reviews posted. I will remove these rows.

```python
nullrows = ulta.isnull().any(axis=1)
ulta.loc[nullrows,:] # all the rows with NAs are rows we want to remove. No missing data otherwise! :D
```

```python
ulta.dropna(axis=0, how='any', inplace = True) # now 653 observations
# ulta.loc[nullrows,:] (re-ran to double check there are no NAs)
```

### Step 2: Data Cleaning
Clean up the entries in each col.

```python
ulta['n_review'] = [s.strip() for s in ulta['n_review']]
ulta['n_review'] = [s.replace('(','') for s in ulta['n_review']] # remove straggler '(' character
```

```python
ulta['price'] = [s.strip() for s in ulta['price']]
```

- some products have a range of prices. for these, the lower end usually represents a travel-sized version of the product. resultingly, I will be taking the (high-end) price, listed after the dash. 

```python
len(ulta.loc[[str.find('-') > -1 for str in ulta['price']]]) # 35 listings to fix
```

```python
clean_prices = [] # empty list to populate
for pri in ulta['price']:
    if (pri.find('-') > -1): # if there is a dash
        clean_prices.append(pri.split('-')[1].replace('$',"").strip()) # take 2nd entry, remove $ and strip
    else:
        clean_prices.append(pri.replace('$',"").strip()) # remove $ and strip for others too
```

```python
clean_prices = [float(price) for price in clean_prices] # make list (soon to be col) float-type
ulta['price'] = clean_prices # re-assign to price col
```

```python
ulta.rename (columns = {'price':'price_usd'}, inplace = True) # for clarity, since we removed $
```

- Lastly, I have to extract the rating from the avg_review col and make it float type.

```python
ulta['avg_review'].value_counts() # cool to know, but also just checking that nothing here is weird
```

```python
clean_means = [float(rev[0:4]) for rev in ulta['avg_review']] # just elements 0-3 
```

```python
ulta['avg_review'] = clean_means
```

```python
ulta.to_csv('ulta-cleaned.csv')
```
